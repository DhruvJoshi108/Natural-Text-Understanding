```python
# Sentiment Analysis
from transformers import pipeline
```

```python
# Load the spam detection pipeline
spam_classification = pipeline('text-classification',model='philschmid/distilbert-base-multilingual-cased-sentiment')
```

```python
txt = [
    'Congratulations! you have won a 50000 INR Amazon gift card, click here to claim',
    'Hi Satyendra Lets meet today at 5PM',
    'URGENT:Your gmail acct has compromised, click here to secure it'
]
results = spam_classification(txt)
label_mapping = {
    'negative':'Negative',
    'netural':'Netural',
    'positive':'Positive'
}
for result in results:
  label = result['label']
  score = result['score']
  # Use the label directly from the model output as the key
  print(f'Sentiment:{label},Confidence Score:{score:.2f}')
```

# Text Summarization

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
```

```python
# Load a pretrained model & tokinizer
model_name = 'facebook/bart-large-cnn'
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
txt = '''Artificial Intelligence is transforming the way we live and work.
It helps in analyzing large amounts of data quickly and accurately.
AI is widely used in healthcare, education, and business industries.
Machine learning allows systems to improve from experience.
With proper use, AI can increase efficiency and reduce human effort.
However, ethical and responsible use of AI is very important.'''
```

```python
# Tokenize the input text
input = tokenizer.encode('Summaries:'+txt,return_tensors='pt',max_length=1024,truncation = True)
```

```python
# Generate the summary
summary_ids = model.generate(input,max_length=150,num_beams =4,early_stopping=True)
summary = tokenizer.decode(summary_ids[0],skip_special_tokens=True)
print('Summary:',summary)
```

# Text Translation

```python
from transformers import AutoModelForSeq2SeqLM
```

```python
translation_model_name = 'Helsinki-NLP/opus-mt-en-es'
translation_tokenizer = AutoTokenizer.from_pretrained(translation_model_name)
translation_model = AutoModelForSeq2SeqLM.from_pretrained(translation_model_name)

input_text = 'My name is Satyendra & I love playing game'
```

```python
input_ids = translation_tokenizer.encode(input_text,return_tensors='pt')
```

```python
# Generate the translated text
translated_ids = translation_model.generate(input_ids,max_length=150,num_beams=4,early_stopping = True)
translated_text = translation_tokenizer.decode(translated_ids[0],skip_special_tokens=True)
print('Translated Text:',translated_text)
```

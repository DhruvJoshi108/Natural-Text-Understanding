# Load OpenAI API Credentials

```python
import locale
locale.getprefrredencoding = lambda: 'UTF-8'
```

```python
import yaml
import os

if not os.path.exists('api_keys.yml'):
    file_content = """OPENAI_API_KEY: 'your_openai_api_key'
GOOGLE_API_KEY: 'your_google_api_key'
NGROK_AUTH_TOKEN: 'your_ngrok_auth_token'
DEEPSEEK_API_KEY: 'your_deepseek_api_key'
"""
    with open('api_keys.yml', 'w') as f:
        f.write(file_content)
    print("Created api_keys.yml with placeholder values. Please update with your actual keys.")

with open('api_keys.yml','r') as file :
  api_creds = yaml.safe_load(file)
```

```python
api_creds.keys()
```

```python
import os
os.environ['OPENAI_API_KEY'] = api_creds['OPENAI_API_KEY']
```
